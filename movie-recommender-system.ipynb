```python
import ast
def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name']) 
    return L 
```

```python
import ast
def convert3(obj):
    L = []
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i['name']) 
            counter+=1
        else:
            break
    return L 
```

# What is Stemming?

## Stemming is an NLP (Natural Language Processing) technique that reduces a word to its root or stem by removing prefixes and suffixes.

```python
# this will convert the ['loved','love','loving '] to a single stemmed word that is love ['love','love','love']
```

```python
import nltk
# we will use this library for this stemming 
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
ps.stem('loved')
```

```python
ps.stem("loving")
```

```python
new_df['tags'][0]
```

```python
stem('in the 22nd century, a paraplegic marine is dispatched to the moon pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. action adventure fantasy sciencefiction cultureclash future spacewar spacecolony society spacetravel futuristic romance space alien tribe alienplanet cgi marine soldier battle loveaffair antiwar powerrelations mindandsoul 3d samworthington zoesaldana sigourneyweaver jamescameron')
```

```python
# now we will apply this stemming to our tags column
new_df['tags']=new_df['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

# MAIN function


```python
# first we will fetch the movie from our data its index..
# if i sort the similarity array i will lose the movie index so thats the problem now 
# in array we have similarity of any movie with 1st movie then 2nd movie and so on ..
# so if we sort this we will lose which similarity with which movie
# index movie change ho gya
# so we will call  enumerate function  we have created tuples having the index postion stored
#
```

```python
# sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
# we will get the 5 movies with most similarity 
```

```python
def recommend(movie):
    movie_index=new_df[new_df['title'] == movie].index[0]
    distances=similarity[movie_index]
    #distance array h 
    movies_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
        
```

```python
recommend('Batman Begins')
```
